```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_data=pd.read_csv("HW1Data/train_data.csv") 
test_data=pd.read_csv("HW1Data/test_data.csv")
```

```python
#1.train_data
x_train = train_data['x'].to_numpy()
y_train = train_data['y'].to_numpy()
m = x_train.shape[0]
x_train = x_train.reshape(m,1)
y_train = y_train.reshape(m,1)
```

```python
x_test = test_data['x'].to_numpy()
y_test = test_data['y'].to_numpy()
m1 = x_test.shape[0]
x_test = x_test.reshape(m1,1)
y_test = y_test.reshape(m1,1)
```

```python
plt.plot(x_train, y_train, 'r.');
plt.plot(x_test, y_test, 'r.');
```

```python
def find_theta(x, y):  
    # Appending a cloumn of ones in X to add the bias term.
    x = np.append(x, np.ones((m,1)), axis=1)
    # The Normal Equation
    theta = np.dot(np.linalg.inv(np.dot(x.T, x)), np.dot(x.T, y))
    return theta
```

```python
def predict(x,y,theta):
    # Appending a cloumn of ones in X to add the bias term.
    x = np.append(x, np.ones((x.shape[0],1)), axis=1)
    # preds is y_hat which is the dot product of X and theta.
    preds = np.dot(x, theta)
    return preds
```

```python
def error(y,pred):
    # Error calculation using MSE
    result = y - pred
    result_sqr = result * result
    mse = np.mean(result_sqr)
    rmse = np.sqrt(mse)
    return rmse
```

```python
# Getting the Value of theta using the find_theta function.
theta = find_theta(x_train, y_train)
theta
pred_train = predict(x_train,y_train,theta)
pred_test = predict(x_test,y_test,theta)
```

```python
# Plotting the predictions.
fig = plt.figure(figsize=(8,6))
plt.plot(x_train, y_train, 'b.')
plt.plot(x_train, pred_train, 'c-')
plt.xlabel('X - Input')
plt.ylabel('y - target / true')
```

```python
fig = plt.figure(figsize=(8,6))
plt.plot(x_test, y_test, 'b.')
plt.plot(x_test, pred_test, 'c-')
plt.xlabel('X - Input')
plt.ylabel('y - target / true')
```

```python
error(y_train,pred_train)
error(y_test,pred_test)
```

```python
def find_theta(x, y,n):
    # Appending a cloumn of ones in X to add the bias term.
    x1 = np.append(x, np.ones((x.shape[0],1)), axis=1)
    for i in range(2,n+1):
        print(i)
        x1 = np.append(x**i, x1, axis=1)
    # The Normal Equation
    theta = np.dot(np.linalg.inv(np.dot(x1.T, x1)), np.dot(x1.T, y))
    return theta
```

```python
def predict(x,y,theta,n):
    x1 = np.append(x, np.ones((x.shape[0],1)), axis=1)
    for i in range(2,n+1):
        print(i)
        x1 = np.append(x**i, x1, axis=1)   
    # preds is y_hat which is the dot product of X and theta.
    preds = np.dot(x1, theta)
    return preds
```

```python
# Getting the Value of theta using the find_theta function.
theta = find_theta(x_train, y_train,5)
theta
pred_train = predict(x_train,y_train,theta,5)
pred_test = predict(x_test,y_test,theta,5)
```

```python
# Plotting the predictions.
fig = plt.figure(figsize=(8,6))
plt.plot(x_train, y_train, 'b.')
plt.plot(x_train, pred_train, 'c-')
plt.xlabel('x_train')
plt.ylabel('y_train')
```
